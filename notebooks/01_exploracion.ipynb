# Exploración de Datasets de GameIQ
Este notebook realiza un análisis exploratorio inicial de los 5 datasets principales.

```python
import pandas as pd
import os
data_path = '../data/'
summary_data = []
```

```python
# --- Carga de VGSales.csv ---
try:
    df = pd.read_csv(os.path.join(data_path, 'VGSales.csv'), encoding='utf-8')
except UnicodeDecodeError:
    df = pd.read_csv(os.path.join(data_path, 'VGSales.csv'), encoding='latin-1')

print(f"Dataset: VGSales.csv")
print(f"Shape: {df.shape}")
print("\nPrimeros 3 registros:")
display(df.head(3))
print("\nCantidad de nulos por columna:")
display(df.isnull().sum())
print("\nTipos de datos:")
display(df.dtypes)
print(f"\nColumna llave identificada: Name")

# Rango temporal
time_info = 'N/A'
try:
    if 'Year' in df.columns:
        if 'date' in 'Year'.lower():
            df['temp_date'] = pd.to_datetime(df['Year'], errors='coerce')
            time_info = f"{df['temp_date'].min().year} - {df['temp_date'].max().year}"
        else:
            time_info = f"{int(df['Year'].dropna().min())} - {int(df['Year'].dropna().max())}"
except:
    pass

summary_data.append({
    'Nombre': 'VGSales.csv',
    'Filas': df.shape[0],
    'Columnas': df.shape[1],
    'Columna_Llave': 'Name',
    'Rango_Temporal': time_info
})
```
